```python
import pandas
```

```python
df=pandas.read_csv("car.data",header=None)
```

```python
df.columns=['buying','maint','doors','persons','lug_boot','safety','target']
```

```python
df
```

```python
Y=df.iloc[:,-1].values
```

```python
X=df.iloc[:,:-1].values
```

```python
from sklearn.preprocessing import LabelEncoder
```

```python
le_1=LabelEncoder()
le_2=LabelEncoder()
le_3=LabelEncoder()
le_4=LabelEncoder()
le_5=LabelEncoder()
le_6=LabelEncoder()
```

```python
X[:,0]=le_1.fit_transform(X[:,0])
X[:,1]=le_2.fit_transform(X[:,1])
X[:,2]=le_3.fit_transform(X[:,2])
X[:,3]=le_4.fit_transform(X[:,3])
X[:,4]=le_5.fit_transform(X[:,4])
X[:,5]=le_6.fit_transform(X[:,5])
```

```python
X
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train,X_test,Y_train,Y_test=train_test_split(X,Y,test_size=0.25)
```

```python
X_train.shape
```

```python
#algorithm 1 knn
```

```python
from sklearn.preprocessing import StandardScaler
sc=StandardScaler()
```

```python
X_train_sc=sc.fit_transform(X_train)
X_test_sc=sc.transform(X_test)
```

```python
from sklearn.neighbors import KNeighborsClassifier
classifier=KNeighborsClassifier(n_neighbors=5,
                                metric='minkowski',
                                p=2)
```

```python
classifier.fit(X_train_sc,Y_train)
```

```python
Y_pred = classifier.predict(X_test_sc)
```

```python
from sklearn.metrics import confusion_matrix
cm = confusion_matrix(Y_test, Y_pred)
```

```python
cm
```

```python
from sklearn.metrics import accuracy_score
accuracy_score(Y_pred,Y_test)
```

```python
#algorithm decision tree
```

```python
from sklearn.tree import DecisionTreeClassifier
classifier=DecisionTreeClassifier(criterion='entropy',random_state=0)
classifier.fit(X_train,Y_train)
```

```python
Y_pred=classifier.predict(X_test)
cm=confusion_matrix(Y_test,Y_pred)
```

```python
accuracy_score(Y_pred,Y_test)
```
